```python
import cv2
from ultralytics import solutions
```

```python
video_path = "/content/drive/MyDrive/car20.mp4"
output_path = "/content/object_counting_output.avi"
```

```python
cap = cv2.VideoCapture(video_path)
assert cap.isOpened(), "Error reading video file"
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python

w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
```

```python

# تعریف ناحیه شمارش (Polygon Region)
region_points = [(20, 400), (1080, 400), (1080, 360), (20, 360), (20, 400)]
```

```python

# تنظیمات ذخیره ویدیو خروجی
video_writer = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))
```

```python

# بارگذاری مدل و تنظیم شمارش اشیا
counter = solutions.ObjectCounter(
    show=False,  # چون در Colab نمی‌توان تصویر زنده نمایش داد، آن را False بگذارید
    region=region_points,
)
```

```python

# پردازش فریم‌ها
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video processing completed.")
        break
    im0 = counter.count(im0)
    video_writer.write(im0)
```

```python
cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
